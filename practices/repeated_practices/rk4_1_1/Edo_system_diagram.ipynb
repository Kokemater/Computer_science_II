```python
import numpy as np
import matplotlib.pyplot as plt

def flujo_sistema(A, xlim=(-10, 10), ylim=(-1000, 1000), densidad=40):
    """
    Dibuja el diagrama de flujo de un sistema de EDOs con coeficientes constantes.
    
    Args:
        A (np.ndarray): Matriz de coeficientes del sistema.
        xlim (tuple): Lmites del eje x (min, max).
        ylim (tuple): Lmites del eje y (min, max).
        densidad (int): Nmero de lneas de flujo por unidad de rea.
    """
    # Generar un espacio de puntos en 2D
    x = np.linspace(xlim[0], xlim[1], densidad)
    y = np.linspace(ylim[0], ylim[1], densidad)
    X, Y = np.meshgrid(x, y)
    
    # Evaluar el campo vectorial
    U = A[0, 0] * X + A[0, 1] * Y
    V = A[1, 0] * X + A[1, 1] * Y
    
    # Crear la figura y el diagrama de flujo
    plt.figure(figsize=(8, 6))
    plt.quiver(X, Y, U, V, color="blue", angles="xy")
    plt.title("Diagrama de flujo del sistema $\dot{x} = A x$")
    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
    plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

# Ejemplo de uso
A = np.array([[-2, -2],
              [-2, -2]])  # Matriz de coeficientes
flujo_sistema(A, xlim=(-2, 2), ylim=(-2, 2), densidad=25)
```
