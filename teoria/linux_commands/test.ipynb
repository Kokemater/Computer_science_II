```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,10)
y = 2 * x

def MSE(y,y_pred):
    return np.sum(0.5*(y - y_pred)**2)
a_0 = 0
a_1 = 0
learning_rate = 0.1
errors = []
iterations = list(range(1000))
for i in iterations:
    y_pred = a_0 + a_1*x
    a_0 -= learning_rate * -(y - y_pred)
    a_1 -= learning_rate * -(y - y_pred)*a_1

    errors.append(MSE(y, y_pred))
```

```python
plt.plot(iterations, errors, marker='o')

# Añadir título y etiquetas
plt.title('Gráfico de Líneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Mostrar el gráfico
plt.show()
```

```python
plt.scatter(x, y, marker='o', color='blue', label='Datos de dispersión')
x_line = np.linspace(0,10,10)
y_line = a_0 + a_1 * x_line
# Crear la línea
plt.plot(x_line, y_line, color='red', label='Línea de ajuste')

# Añadir título y etiquetas
plt.title('Gráfico de Líneas y Dispersión')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Añadir una leyenda
plt.legend()

# Mostrar el gráfico
plt.grid(True)  # Opcional: Añadir una cuadrícula para mejor visualización
plt.show()
```
