```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(v):
    P = 3.5e7
    a = 0.401
    N = 1000
    b = 42.7e-6
    k = 1.3806503e-23
    T = 300
    return (P + a * (N/v)**2 *(v - N*b))
def g():
    P = 3.5e7
    a = 0.401
    N = 1000
    b = 42.7e-6
    k = 1.3806503e-23
    T = 300
    return -k*N*T

plt.figure()
x = np.linspace(0.006,0.1,10000)
f1 = f(x)
g1 = g() * np.ones_like(x)
plt.plot(x,f1)
plt.plot(x,g1)
plt.grid()
plt.show()
```
